##### 차대차 사건 중 죽거나 다친 사람이 많은 발생지 시도별 데이터를 시각화 해보자

```python
# **여기 시험 많이 나옴
# 1. 사건유형_대분류 중 차대차 사건만 뽑아오기

# 2. 해당하는 데이터를 발생지시도별로 합계를 구하기

# 3. x축과 y축 설정하기

# 4. 그래프 그리기
```

```python
# 1. 사고유형_대분류 중 차대차 사건만 뽑아오기
data2017.columns
```

```python
data2017['사고유형_대분류'].unique()
```

```python
ctoc = data2017[data2017['사고유형_대분류'] == '차대차']
ctoc
```

```python
# 2. 해당하는 데이터(차대차 사고)를 발생지 시도별로 합계를 구하기
ctoc_sum = ctoc[['사상자수','발생지시도']].groupby('발생지시도').sum()
ctoc_sum 
```

```python
# 3. x축과 y축 설정하기
x = ctoc_sum.index   # 인덱스 가져오기 .index
y = ctoc_sum.values     # values 값 가져오기 .values
```

```python
# 4. 그래프 그리기
# plot은 꺾은선그래프를 가져오는 친구임

# 각 y축 글씨의 간격 늘리기
plt.figure(figsize = ( 10,8 ))   # 먼저 지정이 필요

plt.plot(x,y,
         marker = 'o',
         c = 'y',
         ls = '--')
plt.xlabel('발생지시도')
plt.ylabel('사상자수', rotation = 0)
plt.title('차대차 교통사고의 사상자 수')
plt.grid()



plt.show()

# 이대로 bar차트를 만드려고하면 만들어지 않음!
```

```python
# y가 지금 2차원의 형태니까 안만들어지는거임
y
```

```python
# 그냥은 바차트가 안만들어지니까 변형시켜서 만드려면
# 1차원으로 변형 시켜줘야함

y_bar = []
for i in range(len(y)) :
    y_bar.append(y[i][0])  # 변형전 y의 배열형태가 한 행에 하나씩
                            # 0열에 들어있으니까
                            # y_bar라는 리스트 안에다가 y의 값을 넣어줌

plt.bar(x,y_bar)

plt.xlabel('발생지시도')
plt.ylabel('사상자수', rotation = 0)
plt.title('차대차 교통사고의 사상자 수')
plt.grid()


plt.show()
```

### 교통사고가 가장 많이 발생하는 시간대를 알아보고 시각화 해보자
- 총 8개의 시간구간으로 나눠서 파이차트 그려보기




```python
data2017.columns
# 발생년월일시를 가져와야 시간구간 나누기 가능!
```

```python
# 1. 발생 년월일시 컬럼에서 시간데이터만 가져오자

time = data2017['발생년월일시'].astype('int64')  # astype을 사용하는 이유는 뒤에서 알려주겠대
time %= 100   #뒤에있는 일 두개만 가져오려면 100으로 나눠서 나머지를 받아줘야함
time
```

```python
# 2. 시간을 카테고리화 시켜보자.
bins = [-1,2,5,8,11,14,17,20,23] # 슬라이싱의 반대라고 생각하면됨
labels =["0-2","3-5","6-8","9-11","12-14","15-17","18-20","21-23"]

# pd.cut() : (범주화 시켜줄 데이터, 쪼개줄 기준, labels = 범주화된 데이터에 라벨)
result = pd.cut(time,bins,labels = labels )
result
```

```python
# 3. 해당하는 데이터로 파이그래프를 그려보자
# value_count = > 각 시간대별 수세기/ sort_index()는 인덱스 기준 정렬하기
rs_data = result.value_counts().sort_index()
rs_data
```

```python
# 파이그래프 만들기
# rs_data를 가지고 파이차트 만들어줄꺼다 / 라벨은 라벨즈라는 것을 이용할꺼다.
# autopct(자동 퍼센트)는 파이차트가 얼마만큼의 지분을 가지고 있는지 자동으로 퍼센트를 계산해줌
plt.pie(rs_data , labels = labels, autopct = '%1.1f%%')
plt.show()
```
