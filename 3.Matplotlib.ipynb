### Line plot
- 꺽은선 그래프


```python
# plt.plot() : 꺾은선 그래프를 그려주는 차트
# y축만 지정/ x축은 그냥 자동으로 설정되어서 나옴
y = [2,4,6 ]
plt.plot(y)
plt.show
```

```python
x = [1,2,3]  #x축을 지정해주니까 1부터 시작하는 것으로 나옴
y = [ 2,4,6]
plt.plot(x,y)
plt.show
```

```python
# ls = 
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,
        ls = '--')
plt.show
```

```python
#marker
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,
        marker = 'o')
plt.show
```

```python
# lw = : line width (선 그래프의 두께를 설정하는 기능)
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,
        lw = 5)
plt.show
```

```python
# c = : 선의 색깔을 지정해주는 기능
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,
        lw = 5,
        c = '#FF00FF')
plt.show
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y,
        lw = 5,
        c = '#FF00FF',
        marker = 'D',
        ms = 15,      # ms : 마커의 사이즈
        mec = 'b',    # mec : 마커 테두리 색상
        mew = 7,       # mew : 마커 테두리 굵기
        mfc = 'r'     # mfc : 마커 내부의 색상
        )
plt.show
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y,
        ls = '--',
        c = 'g',
        marker = 'o'
        )
plt.xlim(-2,10)  # x축의 범위 정해주기
plt.ylim(-5,12)  # y축의 범위 정해주기 
plt.grid()   # 틱값을 기준으로 그래프 평면상에 모눈종이를 그려준다!

plt.show
```

```python
x = np.arange(8)
y = [5,2,3,6,8,7,5,7]

plt.plot(x,y,
        ls = ':',
        c = 'r',
        marker = 'd',
        mfc = 'b'
        )
plt.show
```

```python
# plt.show()이 무엇인지 알아보자!
# 여러 개의 plot 그려보기
x = [ 1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x)
plt.plot(y)

plt.show() # 그래프를 그리고 나눠주는 분기점이라고 생각하면됨

plt.plot(z)

plt.show()   # 도화지! 그래프를 그리고 나눠주는 분기점이라고 생각하면됨
```

```python
# 범례만들기
# 이름설정
x = [ 1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x,label = 'x')  # label = '그래프에 붙여주고 싶은 간단한 명칭'
plt.plot(y, label = 'y')
plt.plot(z, label = 'z')

plt.legend()        # 범례 출력
plt.title('x-y-z')  # 그래프 이름 설정
plt.xlabel('X')     # x라벨(x축 이름 설정)
plt.ylabel('Y',rotation = 0)     # y라벨(y축 이름 설정)
                # rotation = 각도 를 넣어주면 그만큼 회전함
                # 0을 넣어주면 돌지말고 다시 제자리로 돌아오라는 뜻

plt.show()
```
