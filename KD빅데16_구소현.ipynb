### Question3

```python
import pandas as pd
```

```python
data2017 = pd.read_csv('./data/Traffic_Accident_2017.csv',encoding= 'EUC-KR')
data2017
```

```python
data2017.columns
```

```python
# 조건 1. 사고유형_대분류가 ‘차대사람’ 이면서 발생시간대가 ‘야간’인 경우의 사고 건수 카운트
ctop= data2017[data2017['사고유형_대분류'] == '차대사람']
```

```python
night = ctop[ctop['주야'] == '야간']
night['발생년월일시'].count()
```

```python
# 요일별 야간 차대 사람 교통사고 건수
temp = night.loc[ : , '요일'].value_counts()
y = temp[ ['월','화','수','목','금','토','일']]  
y
```

```python
x = y.index

# 조건 3. 차트 내부 글씨는 한글로 표시할 것
from matplotlib import rc
rc('font', family = "Malgun Gothic")

# 조건 2. y축의 범위는 0~200으로 설정할 것
plt.ylim(0,200)

plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대사람 교통사고 건수')


plt.show()
```
